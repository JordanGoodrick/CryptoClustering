```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Import the StandardScaler
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Select the features you want to normalize
features = df_market_data.select_dtypes(include=['float64', 'int64']).columns

# Normalize the data
df_normalized = df_market_data[features].copy()
df_normalized[features] = scaler.fit_transform(df_normalized[features])

# Display the normalized data
df_normalized.head(10)
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))

# Display the list of k-values
k_values
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    # Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=42)  # Set random_state for reproducibility
    
    
    kmeans.fit(df_normalized)  # Ensure you're using the normalized DataFrame
    
    # Append the model's inertia to the inertia_values list
    inertia_values.append(kmeans.inertia_)

# Display the inertia values
inertia_values
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    'k': k_values,  # List of k-values (1 to 11)
    'inertia': inertia_values  # Corresponding inertia values
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Display the DataFrame
df_elbow.head(10)
```

```python
# Plot a line chart with all the inertia values computed for different values of k
df_elbow.hvplot.line(
    x='k', 
    y='inertia', 
    title='Elbow Curve', 
    xlabel='Number of Clusters (k)', 
    ylabel='Inertia',
    line_width=2,
    width=800,
    height=400
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4, as that is where the elbow is. 

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model with k=4
kmeans_best = KMeans(n_clusters=4, random_state=42)

# Fit the model to the scaled data
kmeans_best.fit(df_normalized)

# Display the cluster centers and inertia
kmeans_best.cluster_centers_, kmeans_best.inertia_
```

```python
# Initialize the K-Means model with k=4 (best value for k from the Elbow method)
kmeans_best = KMeans(n_clusters=4, random_state=42)

# Fit the K-Means model using the scaled DataFrame (df_normalized)
kmeans_best.fit(df_normalized)

# Display the cluster centers and inertia
print("Cluster Centers:\n", kmeans_best.cluster_centers_)
print("Inertia:", kmeans_best.inertia_)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
cluster_predictions = kmeans_best.predict(df_normalized)

# Print the resulting array of cluster values
print("Cluster Assignments for each cryptocurrency:\n", cluster_predictions)
```

```python
# Create a copy of the scaled DataFrame
df_scaled_copy = df_normalized.copy()

# Display the first 5 rows of the copied DataFrame
df_scaled_copy.head()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_scaled_copy['cluster'] = cluster_predictions

# Display the copy of the scaled DataFrame with the new column
df_scaled_copy.head()
```

```python
# Create a scatter plot using hvPlot
df_scaled_copy.hvplot.scatter(
    x="price_change_percentage_24h", 
    y="price_change_percentage_7d", 
    c="cluster",  
    cmap='viridis',  
    hover_cols=["coin_id"],  
    title="Cryptocurrency Clusters Based on Price Change",
    xlabel="Price Change Percentage (24h)",
    ylabel="Price Change Percentage (7d)",
    width=800,
    height=400
)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with fit_transform to reduce the original scaled DataFrame to three principal components
pca_data = pca.fit_transform(df_normalized)

# Convert the PCA data into a DataFrame
df_pca = pd.DataFrame(pca_data, columns=['PCA1', 'PCA2', 'PCA3'])

# Display the scaled PCA data
df_pca.head()
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component
explained_variance = pca.explained_variance_ratio_

# Display the explained variance for each principal component
explained_variance

```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 0.8950 (89.5%)

```python
# Create a new DataFrame with the PCA data
df_pca_with_names = pd.DataFrame(pca_data, columns=['PCA1', 'PCA2', 'PCA3'])

# Copy the crypto names (coin_id) from the original DataFrame
df_pca_with_names['coin_id'] = df_scaled_copy.index

# Set the coin_id column as the index
df_pca_with_names.set_index('coin_id', inplace=True)

# Display the scaled PCA DataFrame
df_pca_with_names.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))

# Display the list of k-values
print(k_values)
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
for k in range(1, 12):  # Loop through k-values from 1 to 11
    # Create a KMeans model using the loop counter for n_clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    
    # Fit the model to the data using df_market_data_pca (PCA transformed data)
    kmeans.fit(df_pca_with_names[['PCA1', 'PCA2', 'PCA3']])
    
    # Append the model's inertia to the inertia_values list
    inertia_values.append(kmeans.inertia_)

# Display the inertia values
print(inertia_values)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    'k': list(range(1, 12)),  # k-values from 1 to 11
    'inertia': inertia_values  # The corresponding inertia values
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Display the DataFrame 
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values to visually identify the optimal value for k
df_elbow.hvplot.line(
    x='k', 
    y='inertia', 
    title="Elbow Curve", 
    xlabel="Number of Clusters (k)", 
    ylabel="Inertia",
    width=800, 
    height=400
)
```

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
# Initialize the K-Means model using the best value for k (k = 4)
kmeans_best = KMeans(n_clusters=4, random_state=42)

# Display the initialized model
print(kmeans_best)
```
